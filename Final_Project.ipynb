# Dataset Summary

The Sales & Production Dataset provides comprehensive transactional and operational data for a retail business. It includes detailed information about customers, orders, order items, products, categories, brands, stores, staff, and inventory levels.

The dataset represents the complete sales lifecycle, starting from customer orders through product sales and store fulfillment. In addition to sales data, it contains production-related details such as product categorization, branding, and stock quantities across multiple stores.

This dataset enables in-depth analysis of sales performance, customer purchasing behavior, product demand, store efficiency, and inventory management. It is well-suited for business intelligence and data analytics applications, supporting data-driven decision-making to optimize sales strategies, manage inventory effectively, and improve overall business performance.

# Dataset Overview

The dataset consists of multiple related tables that together describe the sales and production operations of a retail business. It includes both transactional and master data covering customers, orders, products, stores, staff, and inventory.

Overall, the dataset contains a mix of **numerical** and **categorical** features that support detailed analysis of sales performance, customer behavior, product demand, and inventory management.

| Column Type                | Description                                                                 |
|----------------------------|-----------------------------------------------------------------------------|
| Numerical (int64, float64) | IDs, quantities, prices, discounts, stock quantities, and sales values.    |
| Categorical (object)       | Customer details, product names, categories, brands, store locations, dates |

**Data types include:**

- **int64**: Identifiers and count-based values (e.g., IDs, quantities).
- **float64**: Monetary and percentage values (e.g., prices, discounts).
- **object (categorical/text)**: Names, locations, dates, and descriptive fields.

This structure enables efficient use of SQL, Python, and Business Intelligence tools for end-to-end data analysis and visualization.

## Numerical Feature Insights

* **Quantity (Order Items)**:  
The average quantity sold per order item reflects typical customer purchasing behavior, with most orders containing a small to moderate number of units. This indicates that customers generally buy products for personal use rather than bulk purchasing, while occasional higher quantities highlight promotional or high-demand products.

* **List Price**:  
Product prices vary significantly across categories and brands, indicating a diverse product portfolio that includes both affordable and premium items. This price variation supports different customer segments and purchasing power levels.

* **Discount**:  
Discount values are generally low, suggesting that most sales occur at or near the listed price. However, higher discount values in certain transactions may indicate promotional campaigns or clearance sales aimed at boosting demand.

* **Sales Value (Quantity  List Price)**:  
Total sales per order item show a wide range, reflecting differences in product prices and quantities purchased. High-value transactions are typically driven by either premium products or higher quantities, contributing significantly to overall revenue.

* **Stock Quantity (Inventory Levels)**:  
Inventory quantities vary across stores and products, highlighting differences in demand and supply management. Products with consistently low stock levels may indicate high demand or potential inventory replenishment issues.

* **Order Status (Numeric Representation)**:  
Order status values help distinguish completed sales from pending or rejected orders. Completed orders represent actual revenue, while non-completed orders can provide insights into operational inefficiencies or stock availability issues.

* **Model Year**:  
The model year of products allows analysis of demand for newer versus older models. Newer models tend to have higher sales value, reflecting customer preference for updated or latest products.

Overall, numerical features in this dataset provide strong insights into **sales performance, pricing strategies, customer demand, and inventory management**, enabling effective business analysis and data-driven decision-making.
